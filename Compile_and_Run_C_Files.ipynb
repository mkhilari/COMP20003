# Shells 
A shell is a user interface that allows for access to an operating system. Some shells include 
* *PowerShell* - native to the Windows operating system 
* *Ubuntu Bash* - native to the Linux operating system 

# Command Lines
A command line takes commands as input, and are used in shells. 

Commands have the form 

```
prompt command arg1 arg2 argn 
```

* `prompt` - generated by the command line to give context 
* `command` - provided by the user 
* `arg1 arg2 ... argn` - optional arguments 

Some commands include 
* `pwd` - print current working directory (folder) 
* `cd` - change directory 
* `ls` - list files and folders in directory

# Compiling 
Compiling is the process of taking a source file, checking for errors, and creating an object file that can be run. 

# Compiling and Running C Files 
To compile and run a C file using *GCC* (compiler), 

In a shell, navigate to the directory of the C file. 

Enter command 

```
gcc source_file.c -o object_file -Wall -Werror 
```

This compiles `source_file.c` and creates `object_file`. 

`-Wall` displays all warnings 

`-Werror` makes all warnings errors 

To run the compiled file with optional arguments `arg1 arg2 ... argn`, enter command 

```
./object_file arg1 arg2 ... argn 
```
